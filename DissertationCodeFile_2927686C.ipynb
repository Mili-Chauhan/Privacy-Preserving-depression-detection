```python
#Function to load audio data from a directory
def load_audio_from_directory(directory):
    audio_data = []
    filenames = []
    for filename in os.listdir(directory):
        if filename.endswith('.wav'):
            file_path = os.path.join(directory, filename)
            audio, sr = librosa.load(file_path)
            audio_data.append((audio, sr))
            filenames.append(filename)
    return audio_data, filenames
```

```python
# Function to extract demographics-age,gender,C and P details from data
def extract_demographics(filenames):
    demographics = []
    for filename in filenames:
        match = re.match(r'(\d+)_([PCX])([MF])(\d{2})_(\d).wav', filename)
        if match:
            speaker_id, condition, gender, age, education = match.groups()
            demographics.append({
                'Speaker_ID': speaker_id,
                'Condition': condition,
                'Gender': gender,
                'Age': int(age),
                'Education': int(education)
            })
    return pd.DataFrame(demographics)

# Extracting demographics from filenames
demographics_HC_reading_task = extract_demographics(filenames_HC_reading_task)
demographics_PT_reading_task = extract_demographics(filenames_PT_reading_task)
demographics_HC_interview_task = extract_demographics(filenames_HC_interview_task)
demographics_PT_interview_task = extract_demographics(filenames_PT_interview_task)

# Combining demographics data
demographics_df = pd.concat([
    demographics_HC_reading_task,
    demographics_PT_reading_task,
    demographics_HC_interview_task,
    demographics_PT_interview_task
], ignore_index=True)

# Demographics data plot
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.countplot(data=demographics_df, x='Condition')
plt.title('Condition Distribution')

plt.subplot(1, 3, 2)
sns.countplot(data=demographics_df, x='Gender', hue='Condition')
plt.title('Gender Distribution by Condition')

plt.subplot(1, 3, 3)
sns.histplot(data=demographics_df, x='Age', hue='Condition', bins=10, kde=True)
plt.title('Age Distribution by Condition')

plt.tight_layout()
plt.show()
```

```python
# Function to load features data from a specific CSV file and add an id column
def load_data_from_file(file_path, condition):
    df = pd.read_csv(file_path)
    # Extracting id from the filename
    base_name = os.path.basename(file_path)
    nn = base_name.split('_')[0]
    identifier = f"{nn}_{condition}"
    # identifier = f"{nn}_{ord(condition)}"
    df['id'] = identifier  # Adding identifier column
    return df

# reading task data directory path
folder_hc_reading_task = '/content/drive/MyDrive/DissertationMasterProject/DissertationMasterProject/Androids-Corpus/Reading-Task/audio/HC'
folder_pt_reading_task = '/content/drive/MyDrive/DissertationMasterProject/DissertationMasterProject/Androids-Corpus/Reading-Task/audio/PT'

# Load all data from the HC and PT folders
all_data = []

for file_name in os.listdir(folder_hc_reading_task):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_hc_reading_task, file_name)
        df = load_data_from_file(file_path, 'C')  # 'C' for control
        df['label'] = 0  # Adding label 0 for HC
        all_data.append(df)

for file_name in os.listdir(folder_pt_reading_task):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_pt_reading_task, file_name)
        df = load_data_from_file(file_path, 'P')  # 'P' for patient
        df['label'] = 1  # Adding label 1 for PT
        all_data.append(df)

combined_df = pd.concat(all_data, ignore_index=True)
print(combined_df.shape)
print(combined_df.columns.tolist())
print(combined_df['id'].value_counts())
print(combined_df['id'].head())
```

```python
def convert_df_to_3d_array(df):
    # Group by 'id' to collect all time steps for each sample
    grouped = df.groupby('id')

    #maximum number of time steps
    max_time_steps = max(grouped.size())

    #number of features excluding id and label
    num_features = df.shape[1] - 2

    # Initialising the 3D numpy array with zeros
    X = np.zeros((len(grouped), max_time_steps, num_features), dtype=np.float32)

    # Initialising arrays to store id and label separately
    ids = []
    labels = []

    # repeating the last value
    for idx, (name, group) in enumerate(grouped):
        ids.append(name)
        labels.append(group['label'].values[0])

        # Filling NaN values with the last available value in each column
        group_values = group.drop(columns=['id', 'label']).fillna(method='ffill').fillna(0).values

        # Repeat the last value so that the length of group_values matches max_time_steps
        if group_values.shape[0] < max_time_steps:
            repeat_value = group_values[-1]
            padding = np.tile(repeat_value, (max_time_steps - group_values.shape[0], 1))
            group_values = np.vstack((group_values, padding))

        X[idx, :group_values.shape[0], :] = group_values

    return X, np.array(ids), np.array(labels)

# Converting each DataFrame to a 3D array
X_train, train_ids, y_train = convert_df_to_3d_array(train_df)
X_val, val_ids, y_val = convert_df_to_3d_array(val_df)
X_test, test_ids, y_test = convert_df_to_3d_array(test_df)

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("X_train:", X_train)
print("y_train:", y_train)
print("Shape of X_val:", X_val.shape)
print("Shape of y_val:", y_val.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Function for data augumentation
def augment_shift(tensor, shift_amounts):
    augmented_tensors = []
    for shift in shift_amounts:
        shifted_tensor = np.roll(tensor, shift=shift, axis=1)
        augmented_tensors.append(shifted_tensor)

    # Stacking the original and augmented tensors along the first axis
    augmented_tensors = np.concatenate(augmented_tensors, axis=0)
    return augmented_tensors

#shift amounts
shift_amounts = [100, -100, 200, -200]

# augumenting the data
augmented_tensor = augment_shift(X_train, shift_amounts)
augmented_y_train = np.tile(y_train, len(shift_amounts))
X_train = np.concatenate((X_train, augmented_tensor), axis=0)
y_train = np.concatenate((y_train, augmented_y_train), axis=0)

augmented_tensor = augment_shift(X_val, shift_amounts)
augmented_y_val = np.tile(y_val, len(shift_amounts))
X_val = np.concatenate((X_val, augmented_tensor), axis=0)
y_val = np.concatenate((y_val, augmented_y_val), axis=0)

augmented_tensor = augment_shift(X_test, shift_amounts)
augmented_y_test = np.tile(y_test, len(shift_amounts))
X_test = np.concatenate((X_test, augmented_tensor), axis=0)
y_test = np.concatenate((y_test, augmented_y_test), axis=0)

new_list=[]
for i in range(0,len(shift_amounts)):
    for j in train_ids:
        new_list.append(j + '_' + str(i))
train_ids=np.append(train_ids,np.array(new_list))
print(train_ids.shape)

new_list=[]
for i in range(0,len(shift_amounts)):
    for j in val_ids:
        new_list.append(j + '_' + str(i))
val_ids=np.append(val_ids,np.array(new_list))
print(val_ids.shape)

new_list=[]
for i in range(0,len(shift_amounts)):
    for j in test_ids:
        new_list.append(j + '_' + str(i))
test_ids=np.append(test_ids,np.array(new_list))
print(test_ids.shape)

train_labels = np.concatenate((train_labels,np.tile(train_labels, len(shift_amounts))))
val_labels=np.concatenate((val_labels,np.tile(val_labels, len(shift_amounts))))
test_labels=np.concatenate((test_labels,np.tile(test_labels, len(shift_amounts))))


print("Augmented Tensor Shape:", augmented_tensor.shape)
print("Augmented y train Shape: ",augmented_y_train.shape)
print("Combined Tensor Shape:", X_train.shape)
```

```python
#function to convert the dataframe to 3d array
def convert_df_to_3d_array(df, max_time_steps):
    grouped = df.groupby('id')
    num_features = df.shape[1] - 2  # Exclude id and label'

    # Initialising the 3D numpy array with zeros
    X = np.zeros((len(grouped), max_time_steps, num_features), dtype=np.float32)

    ids = []
    labels = []

    for idx, (name, group) in enumerate(grouped):
        ids.append(name)
        labels.append(group['label'].values[0])

        group_values = group.drop(columns=['id', 'label']).fillna(method='ffill').fillna(0).values

        # Repeating the last value so that the length of group_values matches max_time_steps
        if group_values.shape[0] < max_time_steps:
            repeat_value = group_values[-1]
            padding = np.tile(repeat_value, (max_time_steps - group_values.shape[0], 1))
            group_values = np.vstack((group_values, padding))

        X[idx, :group_values.shape[0], :] = group_values

    return X, np.array(ids), np.array(labels)
```

# **Adversarial Model Training: Reading Task**

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Function

# Gradient Reversal Layer
class GradientReversalLayer(Function):
    @staticmethod
    def forward(ctx, x, lambda_):
        ctx.lambda_ = lambda_
        return x

    @staticmethod
    def backward(ctx, grad_output):
        return grad_output.neg() * ctx.lambda_, None

class GradientReversal(nn.Module):
    def __init__(self, lambda_=1.0):
        super(GradientReversal, self).__init__()
        self.lambda_ = lambda_

    def forward(self, x):
        return GradientReversalLayer.apply(x, self.lambda_)

# LSTM Feature Extractor
class LSTMFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTMFeatureExtractor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        h0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)
        c0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        return out[:, -1, :]  # output from the last time step

# Depression Classifier
class DepressionClassifier(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(DepressionClassifier, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc3 = nn.Linear(hidden_size // 2, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return torch.sigmoid(self.fc3(x))


# Gender Classifier
class GenderClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, lambda_):
        super(GenderClassifier, self).__init__()
        self.lambda_ = lambda_
        self.grl = GradientReversal(self.lambda_)
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc3 = nn.Linear(hidden_size // 2, 1)

    def forward(self, x):
        x = self.grl(x)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return torch.sigmoid(self.fc3(x))

# Combine everything into a single model
class AdversarialModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, lambda_):
        super(AdversarialModel, self).__init__()
        self.lambda_ = lambda_
        self.feature_extractor = LSTMFeatureExtractor(input_size, hidden_size, num_layers)
        self.depression_classifier = DepressionClassifier(hidden_size, hidden_size)
        self.gender_classifier = GenderClassifier(hidden_size, hidden_size, self.lambda_)

    def forward(self, x):
        features = self.feature_extractor(x)
        depression_output = self.depression_classifier(features)
        gender_output = self.gender_classifier(features)
        return depression_output, gender_output
```

# **Early Stopping**

```python
#early Stopping Function/Class
class EarlyStopping:
    def __init__(self, patience=10, verbose=False, delta=0):
        self.patience = patience
        self.verbose = verbose
        self.delta = delta
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf

    def __call__(self, val_loss, model):
        score = -val_loss

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose:
                print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    # Function to save model when validation loss decrease.
    def save_checkpoint(self, val_loss, model):
        if self.verbose:
            print(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}).  Saving model ...')
        torch.save(model.state_dict(), 'checkpoint.pt')
        self.val_loss_min = val_loss
```

```python
# Hyperparameter Training Function
def train_and_evaluate(input_size, hidden_size, num_layers, lambda_, dropout_prob, learning_rate, num_epochs, train_loader, val_loader, test_loader, device):
    model = AdversarialModel(input_size, hidden_size, num_layers, lambda_).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    depression_loss_fn = nn.BCELoss()
    gender_loss_fn = nn.BCELoss()

    early_stopping = EarlyStopping(patience=10, verbose=True)

    # Lists to store loss and accuracy
    train_depression_losses = []
    train_gender_losses = []
    train_depression_accuracies = []
    train_gender_accuracies = []

    val_depression_losses = []
    val_gender_losses = []
    val_depression_accuracies = []
    val_gender_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        correct_depression = 0
        correct_gender = 0
        total_samples = 0
        epoch_depression_loss = 0
        epoch_gender_loss = 0

        for batch in train_loader:
            X_batch, y_batch = batch[0], batch[1]
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            # Forward pass
            depression_preds, gender_preds = model(X_batch)

            # Calculate losses
            depression_loss = depression_loss_fn(depression_preds, y_batch.unsqueeze(1))
            pseudo_gender_labels = torch.randint(0, 2, gender_preds.shape).float().to(device)
            gender_loss = gender_loss_fn(gender_preds, pseudo_gender_labels)
            total_loss = depression_loss - lambda_ * gender_loss

            # Backward pass and optimisation
            optimizer.zero_grad()
            total_loss.backward()
            optimizer.step()

            # Accumulate losses
            epoch_depression_loss += depression_loss.item()
            epoch_gender_loss += gender_loss.item()

            # Calculate accuracy
            predicted_depression_labels = (depression_preds >= 0.5).float()
            predicted_gender_labels = (gender_preds >= 0.5).float()

            correct_depression += (predicted_depression_labels == y_batch.unsqueeze(1)).sum().item()
            correct_gender += (predicted_gender_labels == pseudo_gender_labels).sum().item()
            total_samples += y_batch.size(0)

        # Calculate average losses and accuracy for this epoch
        avg_depression_loss = epoch_depression_loss / len(train_loader)
        avg_gender_loss = epoch_gender_loss / len(train_loader)
        train_depression_accuracy = correct_depression / total_samples
        train_gender_accuracy = correct_gender / total_samples

        # Storing the values
        train_depression_losses.append(avg_depression_loss)
        train_gender_losses.append(avg_gender_loss)
        train_depression_accuracies.append(train_depression_accuracy)
        train_gender_accuracies.append(train_gender_accuracy)

        # Validation
        val_results = validate(model, val_loader, depression_loss_fn, gender_loss_fn, lambda_, device)

        # Unpacking the values
        val_avg_depression_loss, val_avg_gender_loss, val_depression_accuracy, val_gender_accuracy, *extra_metrics = val_results
        val_depression_losses.append(val_avg_depression_loss)
        val_gender_losses.append(val_avg_gender_loss)
        val_depression_accuracies.append(val_depression_accuracy)
        val_gender_accuracies.append(val_gender_accuracy)

        # Early stopping
        early_stopping(val_avg_depression_loss, model)
        if early_stopping.early_stop:
            print("Early stopping")
            break

    # Testing
    test_results = test(model, test_loader, depression_loss_fn, gender_loss_fn, lambda_, device)

    test_avg_depression_loss, test_avg_gender_loss, test_depression_accuracy, test_gender_accuracy, *extra_test_metrics = test_results

    return {
        'test_depression_loss': test_avg_depression_loss,
        'test_gender_loss': test_avg_gender_loss,
        'test_depression_accuracy': test_depression_accuracy,
        'test_gender_accuracy': test_gender_accuracy,
    }

```

```python
from itertools import product

# Define hyperparameter grids
hidden_sizes = [64]
num_layers_list = [2]
lambda_list = [0.1,0.5]
dropout_probs = [0.3,0.5]
# learning_rates = [0.009,0.007,0.005]
learning_rates = [0.01,0.005,0.001]
num_epochs = 20

#tracking the best results
best_score = -float('inf')
best_params = None
best_results = None
result_list=[]
# Grid search
for hidden_size, num_layers, lambda_, dropout_prob, learning_rate in product(hidden_sizes, num_layers_list, lambda_list, dropout_probs, learning_rates):
    print(f"Testing params: hidden_size={hidden_size}, num_layers={num_layers}, lambda_={lambda_}, dropout_prob={dropout_prob}, learning_rate={learning_rate}")

    results = train_and_evaluate(
        input_size=input_size,
        hidden_size=hidden_size,
        num_layers=num_layers,
        lambda_=lambda_,
        dropout_prob=dropout_prob,
        learning_rate=learning_rate,
        num_epochs=num_epochs,
        train_loader=train_loader,
        val_loader=val_loader,
        test_loader=test_loader,
        device=device
    )
    result_list.append(results)
    score = results['test_depression_accuracy']
    if score > best_score:
        best_score = score
        best_params = (hidden_size, num_layers, lambda_, dropout_prob, learning_rate)
        best_results = results

print(f"Best Parameters: {best_params}")
print(f"Best Results: {best_results}")
```

```python
# Function to create 3D dataframe
def convert_df_to_3d_array(df):
    # Group by 'id' to collect all time steps for each sample
    grouped = df.groupby('id')

    #maximum number of time steps
    max_time_steps = max(grouped.size())

    #number of features excluding id and label
    num_features = df.shape[1] - 2  # Excluding id and labels

    # Initializing the 3D numpy array with zeros
    X = np.zeros((len(grouped), max_time_steps, num_features), dtype=np.float32)

    # Initializing arrays to store id and labels separately
    ids = []
    labels = []

    # Filling the 3D array with actual data, handling NaNs by repeating the last value
    for idx, (name, group) in enumerate(grouped):
        ids.append(name)
        labels.append(group['label'].values[0])

        # Filling NaN values with the last available value in each column
        group_values = group.drop(columns=['id', 'label']).fillna(method='ffill').fillna(0).values

        # Repeating the last value to ensure the length of group_values matches max_time_steps
        if group_values.shape[0] < max_time_steps:
            repeat_value = group_values[-1]
            padding = np.tile(repeat_value, (max_time_steps - group_values.shape[0], 1))
            group_values = np.vstack((group_values, padding))

        X[idx, :group_values.shape[0], :] = group_values

    return X, np.array(ids), np.array(labels)

# Converting each DataFrame to a 3D array
X_train, train_ids, y_train = convert_df_to_3d_array(train_df)
X_val, val_ids, y_val = convert_df_to_3d_array(val_df)
X_test, test_ids, y_test = convert_df_to_3d_array(test_df)

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("X_train:", X_train)
print("y_train:", y_train)
print("Shape of X_val:", X_val.shape)
print("Shape of y_val:", y_val.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

# **Adversarial Model Training: Interview Task**

```python
# Gradient Reversal Layer
class GradientReversalLayer(Function):
    @staticmethod
    def forward(ctx, x, lambda_):
        ctx.lambda_ = lambda_
        return x

    @staticmethod
    def backward(ctx, grad_output):
        return grad_output.neg() * ctx.lambda_, None

class GradientReversal(nn.Module):
    def __init__(self, lambda_=1.0):
        super(GradientReversal, self).__init__()
        self.lambda_ = lambda_

    def forward(self, x):
        return GradientReversalLayer.apply(x, self.lambda_)

# LSTM Feature Extractor
class LSTMFeatureExtractor(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTMFeatureExtractor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        h0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)
        c0 = torch.zeros(self.lstm.num_layers, x.size(0), self.lstm.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        return out[:, -1, :]  # output from the last time step


# Depression Classifier
class DepressionClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, dropout_prob=0.5):
        super(DepressionClassifier, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout_prob)
        self.fc2 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc3 = nn.Linear(hidden_size // 2, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        return torch.sigmoid(self.fc3(x))

# Gender Classifier
class GenderClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, lambda_, dropout_prob=0.5):
        super(GenderClassifier, self).__init__()
        self.lambda_ = lambda_
        self.grl = GradientReversal(self.lambda_)
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout_prob)
        self.fc2 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc3 = nn.Linear(hidden_size // 2, 1)

    def forward(self, x):
        x = self.grl(x)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        return torch.sigmoid(self.fc3(x))

# Combine everything into a single model
class AdversarialModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, lambda_):
        super(AdversarialModel, self).__init__()
        self.lambda_ = lambda_
        self.feature_extractor = LSTMFeatureExtractor(input_size, hidden_size, num_layers)
        self.depression_classifier = DepressionClassifier(hidden_size, hidden_size)
        self.gender_classifier = GenderClassifier(hidden_size, hidden_size, self.lambda_)

    def forward(self, x):
        features = self.feature_extractor(x)
        depression_output = self.depression_classifier(features)
        gender_output = self.gender_classifier(features)
        return depression_output, gender_output
```

```python
# Plotting Function
def plot_results(data_type, num_epochs, depression_losses, gender_losses, accuracy):
    plt.figure(figsize=(15, 6))

    # Plot Depression Loss
    plt.subplot(1, 3, 1)
    plt.plot(range(num_epochs), depression_losses, marker='o', color='b', label=f'{data_type} Depression Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f"{data_type} Depression Loss Over Epochs")
    plt.legend()

    # Plot Gender Loss
    plt.subplot(1, 3, 2)
    plt.plot(range(num_epochs), gender_losses, marker='o', color='r', label=f'{data_type} Gender Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{data_type} Gender Loss Over Epochs')
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 3, 3)
    plt.plot(range(num_epochs), accuracy, marker='o', color='g', label=f'{data_type} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{data_type} Accuracy Over Epochs')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Validation Function
def validate(model, val_loader, depression_loss_fn, gender_loss_fn, lambda_, device):
    model.eval()
    val_correct_predictions = 0
    val_total_samples = 0
    val_epoch_depression_loss = 0
    val_epoch_gender_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in val_loader:
            X_batch, y_batch = batch[0], batch[1]
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            # Forward pass
            depression_preds, gender_preds = model(X_batch)

            # Calculate losses
            depression_loss = depression_loss_fn(depression_preds, y_batch.unsqueeze(1))

            # Generate random pseudo labels for gender
            pseudo_gender_labels = torch.randint(0, 2, gender_preds.shape).float().to(device)

            # Calculate gender loss
            gender_loss = gender_loss_fn(gender_preds, pseudo_gender_labels)

            # Accumulate losses
            val_epoch_depression_loss += depression_loss.item()
            val_epoch_gender_loss += gender_loss.item()

            # Calculate accuracy
            predicted_labels = (depression_preds >= 0.5).float()
            val_correct_predictions += (predicted_labels == y_batch.unsqueeze(1)).sum().item()
            val_total_samples += y_batch.size(0)

            # Collect predictions and labels for metrics calculation
            all_preds.extend(predicted_labels.cpu().numpy().flatten())
            all_labels.extend(y_batch.cpu().numpy().flatten())

    # Calculate average losses, accuracy, precision, recall, and F1 score
    val_avg_depression_loss = val_epoch_depression_loss / len(val_loader)
    val_avg_gender_loss = val_epoch_gender_loss / len(val_loader)
    val_accuracy = val_correct_predictions / val_total_samples
    val_precision = precision_score(all_labels, all_preds, zero_division=1)
    val_recall = recall_score(all_labels, all_preds, zero_division=1)
    val_f1 = f1_score(all_labels, all_preds, zero_division=1)

    return val_avg_depression_loss, val_avg_gender_loss, val_accuracy, val_precision, val_recall, val_f1
# Test Function
def test(model, test_loader, depression_loss_fn, gender_loss_fn, lambda_, device):
    model.eval()
    test_correct_predictions = 0
    test_total_samples = 0
    test_epoch_depression_loss = 0
    test_epoch_gender_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in test_loader:
            X_batch, y_batch = batch[0], batch[1]
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            # Forward pass
            depression_preds, gender_preds = model(X_batch)

            # Calculate losses
            depression_loss = depression_loss_fn(depression_preds, y_batch.unsqueeze(1))

            # Generate random pseudo labels for gender
            pseudo_gender_labels = torch.randint(0, 2, gender_preds.shape).float().to(device)

            # Calculate gender loss
            gender_loss = gender_loss_fn(gender_preds, pseudo_gender_labels)

            # Accumulate losses
            test_epoch_depression_loss += depression_loss.item()
            test_epoch_gender_loss += gender_loss.item()

            # Calculate accuracy
            predicted_labels = (depression_preds >= 0.5).float()
            test_correct_predictions += (predicted_labels == y_batch.unsqueeze(1)).sum().item()
            test_total_samples += y_batch.size(0)

            # Collect predictions and labels for metrics calculation
            all_preds.extend(predicted_labels.cpu().numpy().flatten())
            all_labels.extend(y_batch.cpu().numpy().flatten())

    # Calculate average losses, accuracy, precision, recall, and F1 score
    test_avg_depression_loss = test_epoch_depression_loss / len(test_loader)
    test_avg_gender_loss = test_epoch_gender_loss / len(test_loader)
    test_accuracy = test_correct_predictions / test_total_samples
    test_precision = precision_score(all_labels, all_preds, zero_division=1)
    test_recall = recall_score(all_labels, all_preds, zero_division=1)
    test_f1 = f1_score(all_labels, all_preds, zero_division=1)

    return test_avg_depression_loss, test_avg_gender_loss, test_accuracy, test_precision, test_recall, test_f1
```

```python
# Early Stopping Class/Function
class EarlyStopping:
    def __init__(self, patience=5, verbose=False, delta=0):
        self.patience = patience
        self.verbose = verbose
        self.delta = delta
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf

    def __call__(self, val_loss, model):
        score = -val_loss

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose:
                print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0
    # Function to save the model when validation loss decrease
    def save_checkpoint(self, val_loss, model):
        if self.verbose:
            print(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}).  Saving model ...')
        torch.save(model.state_dict(), 'checkpoint.pt')
        self.val_loss_min = val_loss
```

```python
# HyperParameter Tuning Function
def train_and_evaluate(input_size, hidden_size, num_layers, lambda_, dropout_prob, learning_rate, num_epochs, train_loader, val_loader, test_loader, device):
    model = AdversarialModel(input_size, hidden_size, num_layers, lambda_).to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    depression_loss_fn = nn.BCELoss()
    gender_loss_fn = nn.BCELoss()

    early_stopping = EarlyStopping(patience=10, verbose=True)

    # Lists to store loss and accuracy
    train_depression_losses = []
    train_gender_losses = []
    train_depression_accuracies = []
    train_gender_accuracies = []

    val_depression_losses = []
    val_gender_losses = []
    val_depression_accuracies = []
    val_gender_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        correct_depression = 0
        correct_gender = 0
        total_samples = 0
        epoch_depression_loss = 0
        epoch_gender_loss = 0

        for batch in train_loader:
            X_batch, y_batch = batch[0], batch[1]
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            # Forward pass
            depression_preds, gender_preds = model(X_batch)

            # Calculate losses
            depression_loss = depression_loss_fn(depression_preds, y_batch.unsqueeze(1))
            pseudo_gender_labels = torch.randint(0, 2, gender_preds.shape).float().to(device)
            gender_loss = gender_loss_fn(gender_preds, pseudo_gender_labels)
            total_loss = depression_loss - lambda_ * gender_loss

            # Backward pass and optimization
            optimizer.zero_grad()
            total_loss.backward()
            optimizer.step()

            # Accumulate losses
            epoch_depression_loss += depression_loss.item()
            epoch_gender_loss += gender_loss.item()

            # Calculate accuracy
            predicted_depression_labels = (depression_preds >= 0.5).float()
            predicted_gender_labels = (gender_preds >= 0.5).float()

            correct_depression += (predicted_depression_labels == y_batch.unsqueeze(1)).sum().item()
            correct_gender += (predicted_gender_labels == pseudo_gender_labels).sum().item()
            total_samples += y_batch.size(0)

        # Calculate average losses and accuracy for this epoch
        avg_depression_loss = epoch_depression_loss / len(train_loader)
        avg_gender_loss = epoch_gender_loss / len(train_loader)
        train_depression_accuracy = correct_depression / total_samples
        train_gender_accuracy = correct_gender / total_samples

        # Storing the values
        train_depression_losses.append(avg_depression_loss)
        train_gender_losses.append(avg_gender_loss)
        train_depression_accuracies.append(train_depression_accuracy)
        train_gender_accuracies.append(train_gender_accuracy)

        # Validation
        val_results = validate(model, val_loader, depression_loss_fn, gender_loss_fn, lambda_, device)

        # Unpacking the values
        val_avg_depression_loss, val_avg_gender_loss, val_depression_accuracy, val_gender_accuracy, *extra_metrics = val_results
        val_depression_losses.append(val_avg_depression_loss)
        val_gender_losses.append(val_avg_gender_loss)
        val_depression_accuracies.append(val_depression_accuracy)
        val_gender_accuracies.append(val_gender_accuracy)

        # Early stopping
        early_stopping(val_avg_depression_loss, model)
        if early_stopping.early_stop:
            print("Early stopping")
            break

    # Testing
    test_results = test(model, test_loader, depression_loss_fn, gender_loss_fn, lambda_, device)

    test_avg_depression_loss, test_avg_gender_loss, test_depression_accuracy, test_gender_accuracy, *extra_test_metrics = test_results

    return {
        'test_depression_loss': test_avg_depression_loss,
        'test_gender_loss': test_avg_gender_loss,
        'test_depression_accuracy': test_depression_accuracy,
        'test_gender_accuracy': test_gender_accuracy,

    }

```
